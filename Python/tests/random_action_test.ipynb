```python
from mlagents_envs.environment import UnityEnvironment
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
```

```python
env = UnityEnvironment(file_name='../scenarios/soccer_continous/unity/SoccerTwos.x86_64', worker_id=1, seed=1, side_channels=[])
```

```python
env.reset()
behaviour_names = list(env.behavior_specs)
```

```python
specs = {}
for name in behaviour_names:
    print(name)
    specs[name] = env.behavior_specs[name]
print(specs)
print(np.sum(specs['SoccerTwos?team=0'].observation_shapes))
```

```python

for key in specs:
    print(key)
    spec = specs[key]
    print('agent IDs: ', spec.)
    print('num obs: ', len(spec.observation_shapes))
    vis_obs = any(len(shape) == 3 for shape in spec.observation_shapes)
    print('is vis: ', vis_obs, '\n')
    print('cont: ', spec.is_action_continuous(), ', disc: ', spec.is_action_discrete())
    
    for index, shape in enumerate(spec.observation_shapes):
        if len(shape) == 3:
            print("Here is the first visual observation")
            plt.imshow(decision_steps.obs[index][0,:,:,:])
            plt.show()

    for index, shape in enumerate(spec.observation_shapes):
        if len(shape) == 1:
            print("First vector observations : ", decision_steps.obs[index][0,:])
```

```python

for name in behaviour_names:
    decision_steps, terminal_steps = env.get_steps(name)
    env.set_actions(name, specs[name].create_random_action(len(decision_steps)))
    print(decision_steps.agent_id)
    obs = decision_steps.obs
        
env.step()
```

```python
for episode in range(10):
    print(episode) # debug
    env.reset()
    for name in behaviour_names:
        decision_steps, terminal_steps = env.get_steps(name)
        env.set_actions(name, specs[name].create_empty_action(len(decision_steps)))
    env.step()
    tracked_agent = -1
    done = False
    episode_rewards = 0
    while not done:
        print('step') # debug
        if tracked_agent == -1 and len(decision_steps) >= 1:
            tracked_agent = decision_steps.agent_id[0] 

        for name in behaviour_names:
            decision_steps, terminal_steps = env.get_steps(name)
            env.set_actions(name, specs[name].create_random_action(len(decision_steps)))
            if tracked_agent in decision_steps: # The agent requested a decision
                episode_rewards += decision_steps[tracked_agent].reward
            if tracked_agent in terminal_steps: # The agent terminated its episode
                episode_rewards += terminal_steps[tracked_agent].reward
                done = True
                break
        env.step()
    print(f"Total rewards for episode {episode} is {episode_rewards}")
```

```python
env.close()
```
